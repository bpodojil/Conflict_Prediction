```python
from tqdm import tqdm
import pandas as pd
import numpy as np
import sklearn
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
```

```python
seed=42
```

```python
df=pd.read_excel ('../data/1a_offset.xlsx',index_col=0,na_values=['..'])
year_dummies = pd.get_dummies(df.year, prefix='year').iloc[:,1:]
region_dummies = pd.get_dummies(df.region, prefix='region').iloc[:,1:]
country_dummies = pd.get_dummies(df.ccode, prefix='country: ').iloc[:,1:]

df = pd.concat([df, year_dummies, region_dummies, country_dummies], axis = 1)
```

```python
#Create testing and training sets
array = df.values

#Create X array
X= array[:,4:]

# Standardizing the features
X = StandardScaler().fit_transform(X)

#Create Y array
Y= array[:,3]
Y=Y.astype('int')

#Training/Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=seed )
#print('Train/Test Split Complete')

#Impliment SMOTE
sm = SMOTE(random_state = seed) 
X_train, Y_train = sm.fit_sample(X_train, Y_train)
#print('oversampling complete')

pca = PCA(.8)
principalComponents = pca.fit_transform(X_train)
print(pca.n_components_)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
svm = SVC(random_state=seed, gamma='scale', C=1)
```

```python
svm.fit(X_train, Y_train)
Y_pred = svm.predict(X_test)
print(recall_score(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
svm = SVC(random_state=seed)
```

```python
svm.fit(X_train, Y_train)
Y_pred = svm.predict(X_test)
print(recall_score(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
recall_results = []
f1_results = []
accuracy_results = []

for kernel in kernels:
    svm = SVC(kernel=kernel)
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    recall=recall_score(Y_test, Y_pred)
    recall_results.append(recall)
    f1=f1_score(Y_test, Y_pred)
    f1_results.append(f1)
    accuracy=accuracy_score(Y_test, Y_pred)
    accuracy_results.append(accuracy)

line1, = plt.plot(kernels, recall_results, 'b', label='Recall')
line2, = plt.plot(kernels, f1_results, 'r', label='F1')
line3, = plt.plot(kernels, accuracy_results, 'g', label='Accuracy')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('Score')
plt.xlabel('kernels')
plt.show()
```

```python
gammas = [.1, .3, .5, .7, .9]
recall_results = []
f1_results = []
accuracy_results = []

for gamma in tqdm(gammas):
    svm = SVC(gamma=gamma)
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    recall=recall_score(Y_test, Y_pred)
    recall_results.append(recall)
    f1=f1_score(Y_test, Y_pred)
    f1_results.append(f1)
    accuracy=accuracy_score(Y_test, Y_pred)
    accuracy_results.append(accuracy)

line1, = plt.plot(gammas, recall_results, 'b', label='Recall')
line2, = plt.plot(gammas, f1_results, 'r', label='F1')
line3, = plt.plot(gammas, accuracy_results, 'g', label='Accuracy')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('Score')
plt.xlabel('gammas')
plt.show()
```

```python
cs = [0.1, 1, 10]
recall_results = []
f1_results = []
accuracy_results = []

for c in cs:
    svm = SVC(C=c)
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    recall=recall_score(Y_test, Y_pred)
    recall_results.append(recall)
    f1=f1_score(Y_test, Y_pred)
    f1_results.append(f1)
    accuracy=accuracy_score(Y_test, Y_pred)
    accuracy_results.append(accuracy)

line1, = plt.plot(cs, recall_results, 'b', label='Recall')
line2, = plt.plot(cs, f1_results, 'r', label='F1')
line3, = plt.plot(cs, accuracy_results, 'g', label='Accuracy')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('Score')
plt.xlabel('cs')
plt.show()
```

```python
degrees = [0, 1, 2, 3, 4, 5, 6]
recall_results = []
f1_results = []
accuracy_results = []

for degree in degrees:
    svm = SVC(kernel='poly',degree=degree)
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    recall=recall_score(Y_test, Y_pred)
    recall_results.append(recall)
    f1=f1_score(Y_test, Y_pred)
    f1_results.append(f1)
    accuracy=accuracy_score(Y_test, Y_pred)
    accuracy_results.append(accuracy)

line1, = plt.plot(degrees, recall_results, 'b', label='Recall')
line2, = plt.plot(degrees, f1_results, 'r', label='F1')
line3, = plt.plot(degrees, accuracy_results, 'g', label='Accuracy')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('Score')
plt.xlabel('degrees')
plt.show()
```
